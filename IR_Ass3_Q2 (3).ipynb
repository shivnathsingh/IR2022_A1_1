# Algorithm to calculate error at a particlar iteration

```python
def error_calculation(prevlist,currlist):
  sum=0
  n=len(prevlist)
  for i in range(0,n):
    sum+=abs(prevlist[i]-currlist[i])
  sum=sum/n
  return sum  
```
